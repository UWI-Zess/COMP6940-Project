```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('./models/data/population_cleaned.csv')
```

```python
df
```

```python
# find the average population for all states for all years in the dataset
# make year a column and change "Description" to "state_name"
population = df.melt(id_vars='Description', var_name='year', value_name='population')
population = population.rename(columns={'Description': 'state_name'})
population['year'] = population['year'].astype(int)
population
```

```python
# remove all the years less than 2000
population = population[population['year'] >= 2000]
population
```

```python
# find the average population for all states for all years in the dataset
average_population = population.groupby('state_name')['population'].mean().reset_index()
average_population
```

```python
# convert population to integer instead of scientific notation
average_population['population'] = average_population['population'].astype(int)
average_population
```

```python
# rename the population column to average_population
average_population = average_population.rename(columns={'population': 'average_population'})
average_population
```

```python
# read in the cleaned hate_crime data
hate_crime = pd.read_csv('./models/data/cleaned_hate_crime_forecasting.csv')
hate_crime
```

```python
# find the number of hate crimes per state
df_state = hate_crime.groupby('state_name').size().reset_index(name='number_of_incidents')
df_state
```

```python
# merge the average population with the number of hate crimes
df_state = pd.merge(df_state, average_population, on='state_name')
df_state
```

```python
# find the number of hate crimes per 100,000 people in each state and add it as a new column
df_state['average_hate_crimes_per_100k'] = df_state['number_of_incidents'] / df_state['average_population'] * 100000
df_state
```

```python
# save the data to a csv file
df_state.to_csv('hate_crime_per_100k.csv', index=False)
```
